# Tahap 07: Analisis Macro vs Weighted Average

Tahap ini merupakan bagian dari pembuktian dan evaluasi lanjutan pasca-pengujian model. Tujuannya adalah untuk membuktikan secara matematis nilai *Macro Average* dan *Weighted Average F1-Score* berdasarkan laporan klasifikasi, serta memberikan justifikasi logis mengenai dampak ketimpangan sebaran data (*imbalanced data*) terhadap metrik evaluasi akhir.

```python
# --- LIBRARIES ---
import pandas as pd

# --- 1. INISIALISASI DATA KINERJA MODEL ---
# Data diekstrak secara manual berdasarkan hasil dari Classification Report
data = {
    'Label': ['Keluhan (0)', 'Saran (1)', 'Pujian (2)'],
    'Precision': [0.90, 0.49, 0.76],
    'Recall': [0.80, 0.64, 0.85],
    'F1-Score': [0.84, 0.56, 0.80],
    'Support': [1965, 536, 496] # Total populasi data: 2997
}

df = pd.DataFrame(data)
total_support = df['Support'].sum()

# --- 2. PERHITUNGAN MACRO AVERAGE ---
# Menghitung rata-rata sederhana dari metrik Precision dan Recall tanpa memandang ukuran kelas
macro_prec = df['Precision'].mean()
macro_rec = df['Recall'].mean()

# Perhitungan F1-Score menggunakan metode harmonic mean dari rata-rata presisi dan recall
macro_f1 = 2 * (macro_prec * macro_rec) / (macro_prec + macro_rec)

# --- 3. PERHITUNGAN WEIGHTED AVERAGE ---
# Menghitung rata-rata dengan mempertimbangkan bobot jumlah data (Support) pada masing-masing kelas
weighted_prec = (df['Precision'] * df['Support']).sum() / total_support
weighted_rec = (df['Recall'] * df['Support']).sum() / total_support

# Perhitungan F1-Score berbobot menggunakan harmonic mean
weighted_f1 = 2 * (weighted_prec * weighted_rec) / (weighted_prec + weighted_rec)

# --- 4. OUTPUT HASIL PERHITUNGAN ---
print("-" * 60)
print("PEMBUKTIAN MATEMATIS CLASSIFICATION REPORT")
print("-" * 60)
print(f"Macro Average F1-Score    : {macro_f1:.2f} (Target: 0.74)")
print(f"Weighted Average F1-Score : {weighted_f1:.2f} (Target: 0.79)")
print("-" * 60)

# --- 5. LOGIKA ANALISIS KETIMPANGAN DATA ---
print("\nANALISIS HASIL EVALUASI:")
print(f"Berdasarkan perhitungan empiris, nilai Weighted Average ({weighted_f1:.2f}) ")
print(f"tercatat lebih tinggi dibandingkan nilai Macro Average ({macro_f1:.2f}).")
print("\nKesimpulan:")
print("Terdapat indikasi ketimpangan proporsi data (Data Imbalance) yang signifikan pada dataset.")
print(f"Hal ini dibuktikan oleh mayoritas sebaran data yang terpusat pada kelas 'Keluhan' ({df.loc[0, 'Support']} sampel).")
print("Model memiliki kecenderungan performa prediktif yang jauh lebih kuat pada kelas mayoritas,")
print("sehingga secara matematis dominasi kelas ini mengangkat keseluruhan nilai rata-rata berbobot (Weighted Average).")
print("=" * 60)
```
