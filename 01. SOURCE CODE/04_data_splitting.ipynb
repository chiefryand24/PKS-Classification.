# Tahap 04: Data Splitting

Tahap ini bertujuan untuk membagi dataset yang telah melalui proses *preprocessing* menjadi dua bagian: data *training* dan data *testing*. 

Pembagian dilakukan dengan proporsi 80:20 menggunakan parameter `stratify` untuk memastikan distribusi kelas label (Pujian, Keluhan, Saran) tetap seimbang dan proporsional pada kedua set data tersebut.

```python
# --- LIBRARIES ---
import os
import pandas as pd
from sklearn.model_selection import train_test_split

# --- 1. KONFIGURASI DIREKTORI & PATH ---
BASE_DIR = r'C:\Users\Asus\Desktop\SKRIPSI\1. MAIN'
INPUT_FILE = 'data_preprocessing_2.0.csv'
FILE_PATH = os.path.join(BASE_DIR, INPUT_FILE)
```

## 4.1 Eksekusi Pembagian Data (*Training* & *Testing*)
Fungsi di bawah ini akan memuat dataset bersih, memisahkan variabel *independent* (fitur teks) dan *dependent* (label kelas), serta mencetak laporan distribusi hasil pembagian untuk keperluan dokumentasi Bab 4.

```python
def execute_final_splitting():
    """
    Mengeksekusi pembagian data menggunakan train_test_split 
    serta mencetak laporan proporsi label untuk dokumentasi.
    """
    # Validasi keberadaan file dataset
    if not os.path.exists(FILE_PATH):
        print(f"Error: File {INPUT_FILE} tidak ditemukan di direktori utama!")
        return None, None, None, None

    # Memuat dataset
    df = pd.read_csv(FILE_PATH)
    
    # Langkah preventif: Membersihkan baris yang kosong atau invalid
    df = df.dropna(subset=['text', 'label_pks'])

    # --- 2. PROSES SPLITTING ---
    # X = Fitur (Teks), y = Target (Label)
    X = df['text']
    y = df['label_pks']

    # Pembagian 80:20 dengan Stratify dan Shuffle
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, 
        test_size=0.20,      # Alokasi 20% untuk data testing
        random_state=42,     # Mengunci hasil pengacakan (reproducibility)
        stratify=y,          # Menjaga proporsi label secara merata
        shuffle=True         # Mengacak urutan data sebelum dibagi
    )

    # --- 3. LAPORAN HASIL (Untuk Dokumentasi Bab 4) ---
    print("="*60)
    print("📋 LAPORAN FINAL DATA SPLITTING")
    print("="*60)
    print(f"Total Baris Data Awal  : {len(df)}")
    print(f"Jumlah Data Training   : {len(X_train)} (80%)")
    print(f"Jumlah Data Testing    : {len(X_test)} (20%)")
    print("-" * 60)
    
    # Kalkulasi persentase distribusi label
    dist_train = y_train.value_counts(normalize=True) * 100
    dist_test = y_test.value_counts(normalize=True) * 100
    
    print("Proporsi Label pada Data TRAINING:")
    for label, perc in dist_train.items():
        print(f" - {label.ljust(10)}: {perc:.2f}% ({y_train.value_counts()[label]} data)")
        
    print("\nProporsi Label pada Data TESTING:")
    for label, perc in dist_test.items():
        print(f" - {label.ljust(10)}: {perc:.2f}% ({y_test.value_counts()[label]} data)")
    print("="*60)

    # --- 4. PENYIMPANAN HASIL ---
    # Menyimpan data ke CSV sebagai arsip pengujian yang tetap
    train_df = pd.DataFrame({'text': X_train, 'label_pks': y_train})
    test_df = pd.DataFrame({'text': X_test, 'label_pks': y_test})
    
    train_path = os.path.join(BASE_DIR, 'data_train_final_2.0.csv')
    test_path = os.path.join(BASE_DIR, 'data_test_final_2.0.csv')
    
    train_df.to_csv(train_path, index=False)
    test_df.to_csv(test_path, index=False)
    
    print("\nData Training & Testing telah berhasil disimpan di folder MAIN!")
    
    return X_train, X_test, y_train, y_test

# Eksekusi fungsi utama
if __name__ == "__main__":
    X_train, X_test, y_train, y_test = execute_final_splitting()
```
