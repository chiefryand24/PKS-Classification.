## 10.1 Distribusi Label (Data Labeling)
Visualisasi diagram batang (*bar chart*) untuk menampilkan proporsi sebaran kelas label (Keluhan, Saran, Pujian) pada dataset awal sebelum dilakukan *preprocessing* dan pemisahan data.

```python
# --- LIBRARIES ---
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- 1. KONFIGURASI DIREKTORI DAN GRAFIK ---
DIR_PATH = r'C:\Users\Asus\Desktop\SKRIPSI\1. MAIN'
FILE_NAME = 'data_labeling.csv'
COL_NAME = 'label_pks'  

# Konfigurasi Teks Grafik
LABEL_X = 'Label'  
LABEL_Y = 'Frekuensi'  
TITLE_TEXT = 'Distribusi Label (Data Labeling)'

# Konfigurasi Ukuran Font (Disusun untuk format Skripsi)
FONT_SIZE_TITLE = 20
FONT_SIZE_AXIS_LABEL = 20
FONT_SIZE_TICKS = 16  
FONT_SIZE_BARS = 16

# Konfigurasi Spasi
LABEL_PAD = 20  
# --------------------------------

try:
    # --- 2. PEMUATAN DATA ---
    df = pd.read_csv(os.path.join(DIR_PATH, FILE_NAME))
    counts = df[COL_NAME].value_counts().reset_index()
    counts.columns = ['cat', 'val']
    
    # Pemetaan Warna Kustom (Matte Colors)
    # Keluhan -> Matte Red, Saran -> Matte Yellow, Pujian -> Matte Green
    color_map = {
        'keluhan': '#D65A5A', 
        'saran': '#E6C95D',   
        'pujian': '#7FB285'   
    }
    
    # Membuat palet warna berdasarkan kategori yang ada
    current_palette = [color_map.get(str(label).lower(), '#808080') for label in counts['cat']]
    total = counts['val'].sum()

    # --- 3. PENGATURAN PLOT ---
    sns.set_theme(style="whitegrid")
    fig, ax = plt.subplots(figsize=(12, 8))
    
    sns.barplot(data=counts, x='cat', y='val', palette=current_palette, edgecolor='black', ax=ax)

    # --- 4. PENERAPAN FORMAT TEKS DAN UKURAN ---
    ax.set_title(f'{TITLE_TEXT} (Total: {total})', fontweight='bold', fontsize=FONT_SIZE_TITLE, pad=30)
    ax.set_xlabel(LABEL_X, fontweight='bold', fontsize=FONT_SIZE_AXIS_LABEL, labelpad=LABEL_PAD)
    ax.set_ylabel(LABEL_Y, fontweight='bold', fontsize=FONT_SIZE_AXIS_LABEL, labelpad=LABEL_PAD)

    # Menyesuaikan ukuran label sumbu (ticks)
    ax.tick_params(axis='both', which='major', labelsize=FONT_SIZE_TICKS)
    plt.xticks(fontweight='bold') 
    plt.yticks(fontweight='bold') 

    # --- 5. PENAMBAHAN ANOTASI ANGKA ---
    for p in ax.patches:
        height = p.get_height()
        if height > 0:
            ax.annotate(f'{int(height)}', 
                        (p.get_x() + p.get_width() / 2., height),
                        ha='center', va='bottom', 
                        xytext=(0, 10), textcoords='offset points', 
                        fontweight='bold', fontsize=FONT_SIZE_BARS)

    # Menyesuaikan batas atas sumbu Y agar angka tidak terpotong
    ax.set_ylim(0, counts['val'].max() * 1.15)
    plt.tight_layout()

    # --- 6. EKSEKUSI PENAMPILAN DAN PENYIMPANAN ---
    print("Membuka jendela visualisasi grafik...")
    plt.show()

    ans = input("Simpan grafik dalam format SVG? (y/n): ").lower().strip()
    if ans == 'y':
        save_path = os.path.join(DIR_PATH, 'gambar8.svg')
        # Memastikan format tersimpan sebagai SVG (vektor)
        fig.savefig(save_path, format='svg', bbox_inches='tight')
        print(f"Sukses: Grafik berhasil disimpan ke {save_path}")

except Exception as e:
    print(f"Error: {e}")
```

## 10.2 Visualisasi Word Cloud: Profanity List
Visualisasi ini bertujuan untuk memberikan gambaran mengenai kata-kata kasar (*profanity*) yang paling dominan dalam dataset. *Word Cloud* dikonfigurasi dengan resolusi tinggi (3200x1600) untuk memastikan kualitas gambar tetap tajam saat dicetak dalam dokumen skripsi.

```python
# --- LIBRARIES ---
import os
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# --- 1. KONFIGURASI DIREKTORI DAN FILE ---
DIR_PATH = r'C:\Users\Asus\Desktop\SKRIPSI\1. MAIN'
FILE_NAME = 'profanity_dict.csv'
COL_NAME = 'profanity_list'  

try:
    # --- 2. PEMUATAN DATA ---
    df = pd.read_csv(os.path.join(DIR_PATH, FILE_NAME))
    
    if COL_NAME not in df.columns:
        print(f"Error: Kolom '{COL_NAME}' tidak ditemukan dalam file CSV.")
    else:
        # Membersihkan data dari nilai kosong (NaN)
        clean_df = df[df[COL_NAME].notna()]
        text_data = " ".join(clean_df[COL_NAME].astype(str))

        # --- 3. GENERASI WORD CLOUD (RESOLUSI TINGGI) ---
        # Konfigurasi 3200x1600 untuk memastikan ketajaman visual
        wc = WordCloud(
            width=3200,  
            height=1600, 
            background_color='white',
            colormap='plasma', 
            collocations=False
        ).generate(text_data)

        # --- 4. PENAMPILAN GRAFIK ---
        plt.figure(figsize=(15, 7.5))
        plt.imshow(wc, interpolation='bilinear')
        plt.axis('off')
        plt.title(f"Word Cloud of {COL_NAME}", fontsize=18, fontweight='bold', pad=20)
        plt.tight_layout()
        plt.show()

        # --- 5. LOGIKA PENYIMPANAN (VERSI HD) ---
        ans = input("Simpan grafik Word Cloud versi HD? (y/n): ").lower().strip()
        if ans == 'y':
            save_path = os.path.join(DIR_PATH, 'gambar9.png')
            # Menggunakan metode to_file untuk mempertahankan resolusi asli 3200x1600
            wc.to_file(save_path)
            print(f"Sukses: Gambar HD berhasil disimpan ke {save_path}")

except Exception as e:
    print(f"Terjadi kesalahan: {e}")
```

## 10.3 Visualisasi Sampel Kamus Emoji
Grafik ini menampilkan representasi sampel emoji dari masing-masing kategori klasifikasi (Pujian, Keluhan, Saran). Visualisasi ini dikonfigurasi secara khusus menggunakan font pendukung emoji (seperti *Segoe UI Emoji*) dan pengaturan ekspor SVG agar karakter emoji tetap terbaca dengan jelas saat dimasukkan ke dalam dokumen Microsoft Word.

```python
# --- LIBRARIES ---
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from matplotlib.lines import Line2D

# --- 1. KONFIGURASI DIREKTORI DAN TEKS ---
DIR_PATH = r'C:\Users\Asus\Desktop\SKRIPSI\1. MAIN'
FILE_NAME = 'emoji_dict.csv'

TITLE_TEXT = 'Emoji Dictionary Samples (Top 3 per Class)'
LABEL_X = 'Kategori Klasifikasi'
FONT_SIZE_TITLE = 20
FONT_SIZE_AXIS_LABEL = 20
FONT_SIZE_TICKS = 16  
FONT_SIZE_LEGEND = 14
LABEL_PAD = 20

# --- KONFIGURASI GLOBAL EXPORT SVG ---
# Memastikan teks dan karakter emoji tidak hilang atau terkonversi menjadi path saat di-export
plt.rcParams['svg.fonttype'] = 'none'

# --- PENANGANAN FONT EMOJI ---
def get_emoji_font():
    """Mencari font sistem yang mendukung rendering karakter emoji."""
    for f in fm.findSystemFonts():
        if 'seguiemj' in f.lower() or 'emoji' in f.lower():
            return f
    return None

emoji_font_path = get_emoji_font()
prop = fm.FontProperties(fname=emoji_font_path) if emoji_font_path else None

try:
    # --- 2. PEMUATAN DAN SAMPLING DATA ---
    df = pd.read_csv(os.path.join(DIR_PATH, FILE_NAME))
    
    # Mengambil 3 sampel teratas untuk setiap kategori klasifikasi
    df_sampled = df.groupby('klasifikasi').head(3).copy()
    
    # Format label tampilan: [Emoji] [Makna]
    df_sampled['display'] = df_sampled['emoji_list'] + "    " + df_sampled['makna']
    
    # Melakukan pengurutan kategori agar visualisasi lebih terorganisir
    df_sampled = df_sampled.sort_values('klasifikasi', ascending=True).reset_index(drop=True)

    # --- 3. PEMETAAN WARNA (MATTE COLOR PALETTE) ---
    color_map = {
        'keluhan': '#D65A5A', # Matte Red
        'saran': '#E6C95D',   # Matte Yellow
        'pujian': '#7FB285'   # Matte Green
    }
    colors = [color_map.get(k.lower(), '#808080') for k in df_sampled['klasifikasi']]

    # --- 4. PEMBUATAN GRAFIK ---
    plt.style.use('seaborn-v0_8-whitegrid')
    fig, ax = plt.subplots(figsize=(12, 8))

    # Menggunakan posisi numerik pada sumbu Y untuk menjamin urutan baris
    y_pos = range(len(df_sampled))
    bars = ax.barh(y_pos, [1]*len(df_sampled), color=colors, edgecolor='black')

    # --- 5. FORMATTING DAN TIPOGRAFI ---
    ax.set_title(TITLE_TEXT, fontsize=FONT_SIZE_TITLE, fontweight='bold', pad=30)
    ax.set_xlabel(LABEL_X, fontsize=FONT_SIZE_AXIS_LABEL, fontweight='bold', labelpad=LABEL_PAD)
    
    # Menetapkan label emoji pada sumbu Y
    ax.set_yticks(y_pos)
    ax.set_yticklabels(df_sampled['display'], fontweight='bold')
    
    # Menerapkan properti font emoji jika ditemukan pada sistem
    if prop:
        for label in ax.get_yticklabels():
            label.set_fontproperties(prop)
            label.set_fontsize(FONT_SIZE_TICKS)
    
    # Menyembunyikan angka pada sumbu X karena hanya bersifat representasi kategori
    ax.set_xticks([])
    
    # Menghilangkan garis pembatas atas dan kanan untuk estetika
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    # --- 6. PENAMBAHAN LEGENDA ---
    legend_elements = [
        Line2D([0], [0], color='#D65A5A', lw=8, label='Keluhan'),
        Line2D([0], [0], color='#E6C95D', lw=8, label='Saran'),
        Line2D([0], [0], color='#7FB285', lw=8, label='Pujian')
    ]
    ax.legend(handles=legend_elements, title="Kategori", 
              fontsize=FONT_SIZE_LEGEND, title_fontsize=FONT_SIZE_LEGEND + 2, 
              loc='lower right', frameon=True)

    plt.tight_layout()
    plt.show()

    # --- 7. LOGIKA PENYIMPANAN SVG ---
    ans = input("Simpan grafik sampel emoji dalam format SVG? (y/n): ").lower().strip()
    if ans == 'y':
        save_path = os.path.join(DIR_PATH, 'gambar10.svg')
        fig.savefig(save_path, format='svg', bbox_inches='tight')
        print(f"Sukses: Grafik sampel emoji berhasil disimpan ke {save_path}")

except Exception as e:
    print(f"Terjadi kesalahan: {e}")
```

## 10.4 Perbandingan Kata Tidak Baku (Slang) ke Baku
Visualisasi ini menampilkan sampel proses normalisasi kata dari bentuk tidak baku (*slang*) ke bentuk baku sesuai dengan kamus leksikon yang digunakan. Grafik didesain secara minimalis dengan format dua arah untuk mempermudah pemahaman mengenai transformasi teks pada tahap *preprocessing*.

```python
# --- LIBRARIES ---
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. KONFIGURASI DIREKTORI DAN FILE ---
DIR_PATH = r'C:\Users\Asus\Desktop\SKRIPSI\1. MAIN'
FILE_NAME = 'slang_dict.csv'

try:
    # --- 2. PEMUATAN DAN FILTERING DATA ---
    df = pd.read_csv(os.path.join(DIR_PATH, FILE_NAME))
    
    # Pengelompokkan berdasarkan huruf awal untuk mendapatkan sampel yang beragam (A-J)
    df['first_letter'] = df['slang_list'].astype(str).str[0].str.upper()
    target_letters = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
    df_filtered = df[df['first_letter'].isin(target_letters)].sort_values(['first_letter', 'slang_list'])
    
    # Mengambil tepat satu representasi kata untuk setiap kategori huruf
    df_sample = df_filtered.groupby('first_letter').head(1).reset_index(drop=True)

    # --- 3. PENGATURAN PLOT ---
    sns.set_theme(style="white")
    fig, ax = plt.subplots(figsize=(10, 8)) 

    y_pos = range(len(df_sample))
    
    # Plotting bar horizontal dengan palet warna standar (Biru untuk Baku, Oranye untuk Slang)
    ax.barh(y_pos, [0.8]*len(df_sample), color='#5b84c1', edgecolor='black', label='Formal (Baku)')
    ax.barh(y_pos, [-0.8]*len(df_sample), color='#e4a169', edgecolor='black', label='Slang (Tidak Baku)')

    # --- 4. PENEMPATAN LABEL TEKS ---
    for i, row in df_sample.iterrows():
        # Menempatkan kata Slang di sisi kiri
        ax.text(-0.85, i, str(row['slang_list']), ha='right', va='center', fontweight='bold', fontsize=12)
        
        # Menempatkan kata Formal di sisi kanan
        ax.text(0.85, i, str(row['baku']), ha='left', va='center', fontweight='bold', fontsize=12)
        
        # Menempatkan simbol transisi di tengah
        ax.text(0, i, '→', ha='center', va='center', fontweight='bold', fontsize=16)

    # --- 5. ESTETIKA DAN FORMATTING ---
    ax.set_title('Slang to Baku Sample Comparison', fontsize=16, fontweight='bold', pad=30)
    
    # Menghilangkan seluruh garis pembatas (spines) untuk tampilan minimalis
    for spine in ax.spines.values():
        spine.set_visible(False)
    
    # Menyembunyikan sumbu X dan Y
    ax.set_yticks([]) 
    ax.set_xticks([]) 
    ax.set_xlim(-2.0, 2.0) 
    
    # Mengatur posisi legenda di bagian bawah grafik
    ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), ncol=2, frameon=False, fontsize=12)

    plt.tight_layout()
    plt.show()

    # --- 6. LOGIKA PENYIMPANAN SVG ---
    ans = input("Simpan grafik perbandingan minimalis ini dalam format SVG? (y/n): ").lower().strip()
    if ans == 'y':
        save_path = os.path.join(DIR_PATH, 'gambar11.svg')
        # Format set secara eksplisit ke SVG untuk menjamin kualitas vektor
        fig.savefig(save_path, format='svg', bbox_inches='tight')
        print(f"Sukses: Grafik berhasil disimpan ke {save_path}")

except Exception as e:
    print(f"Terjadi kesalahan: {e}")
```

## 10.7 Perbandingan Distribusi Label (Data Splitting)
Visualisasi ini membandingkan sebaran frekuensi label pada kumpulan data pelatihan (*training set*) dan data pengujian (*testing set*). Grafik ini berfungsi untuk memvalidasi bahwa distribusi kelas (Keluhan, Saran, Pujian) tetap konsisten setelah dilakukan proses pembagian data menggunakan metode *stratified sampling*.

```python
# --- LIBRARIES ---
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- 1. KONFIGURASI DIREKTORI DAN FILE ---
DIR_PATH = r'C:\Users\Asus\Desktop\SKRIPSI\1. MAIN'
TRAIN_FILE = 'data_train_final_2.0.csv'
TEST_FILE = 'data_test_final_2.0.csv'
TARGET_COL = 'label_pks'

# Pengaturan Teks untuk Dokumentasi Skripsi
PLOT_TITLE = 'Comparison of Label Distribution (Data Splitting)'
X_LABEL = 'Label'
Y_LABEL = 'Frekuensi'

# Konfigurasi Ukuran Font (Konsisten dengan format Skripsi)
FONT_SIZE_TITLE = 20
FONT_SIZE_AXIS_LABEL = 20
FONT_SIZE_TICKS = 16  
FONT_SIZE_BARS = 16

# Konfigurasi Spasi
LABEL_PAD = 20  
# --------------------------------

try:
    # --- 2. PEMUATAN DATA OPTIMAL ---
    # Hanya memuat kolom target untuk efisiensi memori
    train_df = pd.read_csv(os.path.join(DIR_PATH, TRAIN_FILE), usecols=[TARGET_COL])
    test_df = pd.read_csv(os.path.join(DIR_PATH, TEST_FILE), usecols=[TARGET_COL])

    # --- 3. AGREGASI DATA ---
    tr_c = train_df[TARGET_COL].value_counts().reset_index()
    tr_c.columns = ['label', 'count']; tr_c['dataset'] = 'Training'
    
    ts_c = test_df[TARGET_COL].value_counts().reset_index()
    ts_c.columns = ['label', 'count']; ts_c['dataset'] = 'Testing'

    combined = pd.concat([tr_c, ts_c])

    # --- 4. PENGATURAN PLOT ---
    sns.set_theme(style="whitegrid")
    fig, ax = plt.subplots(figsize=(12, 8))

    # Membuat diagram batang berkelompok dengan palet warna 'muted'
    sns.barplot(
        data=combined, 
        x='label', 
        y='count', 
        hue='dataset', 
        palette='muted', 
        edgecolor='black', 
        width=0.7,  # Memberikan jarak antar grup batang
        ax=ax
    )

    # --- 5. FORMATTING TEKS DAN UKURAN FONT ---
    ax.set_title(f'{PLOT_TITLE}\n(Total Train: {len(train_df)} | Total Test: {len(test_df)})', 
                  fontweight='bold', fontsize=FONT_SIZE_TITLE, pad=30)
    ax.set_xlabel(X_LABEL, fontweight='bold', fontsize=FONT_SIZE_AXIS_LABEL, labelpad=LABEL_PAD)
    ax.set_ylabel(Y_LABEL, fontweight='bold', fontsize=FONT_SIZE_AXIS_LABEL, labelpad=LABEL_PAD)

    # Mengatur ukuran dan ketebalan label sumbu (ticks)
    ax.tick_params(axis='both', which='major', labelsize=FONT_SIZE_TICKS)
    plt.xticks(fontweight='bold') 
    plt.yticks(fontweight='bold') 

    # --- 6. PENAMBAHAN ANOTASI NILAI DI ATAS BATANG ---
    for p in ax.patches:
        height = p.get_height()
        if height > 0:
            ax.annotate(f'{int(height)}', 
                        (p.get_x() + p.get_width() / 2., height),
                        ha='center', va='bottom', 
                        xytext=(0, 10), textcoords='offset points', 
                        fontweight='bold', fontsize=FONT_SIZE_BARS)

    # Menyesuaikan batas sumbu Y untuk estetika visual
    ax.set_ylim(0, combined['count'].max() * 1.15)
    plt.tight_layout()

    # --- 7. LOGIKA PENAMPILAN DAN PENYIMPANAN ---
    print("Komputasi selesai. Membuka jendela visualisasi grafik...")
    plt.show()

    ans = input("Simpan grafik perbandingan splitting dalam format SVG? (y/n): ").lower().strip()
    if ans == 'y':
        save_path = os.path.join(DIR_PATH, 'gambar14.svg')
        # Format set secara eksplisit ke SVG untuk kualitas vektor maksimal
        fig.savefig(save_path, format='svg', bbox_inches='tight')
        print(f"Sukses: Grafik berhasil disimpan ke {save_path}")

except Exception as e:
    print(f"Terjadi kesalahan: {e}")
```

## 10.8 Visualisasi 10 Fitur Teratas (TF-IDF)
Grafik ini menampilkan 10 kata dengan nilai rata-rata bobot *Term Frequency-Inverse Document Frequency* (TF-IDF) tertinggi dari seluruh dataset pelatihan. Hal ini bertujuan untuk mengidentifikasi kata-kata yang memiliki tingkat kepentingan atau pengaruh paling besar dalam merepresentasikan informasi tekstual pada penelitian ini.

```python
# --- LIBRARIES ---
import os
import pickle
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- 1. KONFIGURASI DIREKTORI DAN FILE ---
DIR_PATH = r'C:\Users\Asus\Desktop\SKRIPSI\1. MAIN'
MODEL_NAME = 'tfidf_model_2.0.pkl'
DATA_NAME = 'data_train_final_2.0.csv'
TEXT_COL = 'text'

# Konfigurasi Ukuran Font (Konsisten dengan format Skripsi)
FONT_SIZE_TITLE = 20
FONT_SIZE_AXIS_LABEL = 20
FONT_SIZE_TICKS = 16  
LABEL_PAD = 20
# -----------------------------------------------

try:
    # --- 2. PEMUATAN MODEL DAN DATA ---
    with open(os.path.join(DIR_PATH, MODEL_NAME), 'rb') as f:
        tfidf = pickle.load(f)
        
    # Memuat data secara optimal hanya pada kolom teks
    df = pd.read_csv(os.path.join(DIR_PATH, DATA_NAME), usecols=[TEXT_COL])

    # --- 3. TRANSFORMASI DATA & KALKULASI RATA-RATA TF-IDF ---
    tfidf_matrix = tfidf.transform(df[TEXT_COL])
    weights = tfidf_matrix.mean(axis=0).tolist()[0]
    features = tfidf.get_feature_names_out()

    # --- 4. PENYUSUNAN DATAFRAME TOP 10 ---
    df_weights = pd.DataFrame({'word': features, 'weight': weights})
    df_top10 = df_weights.sort_values(by='weight', ascending=False).head(10)

    # --- 5. PEMBUATAN GRAFIK (OBJECT-ORIENTED STYLE) ---
    sns.set_theme(style="whitegrid")
    fig, ax = plt.subplots(figsize=(12, 8))
    
    # Menggunakan palet 'mako' untuk gradasi warna profesional
    sns.barplot(
        data=df_top10, 
        x='weight', 
        y='word', 
        palette='mako', 
        edgecolor='black',
        ax=ax
    )

    # --- 6. FORMATTING TEKS, KETEBALAN, DAN SPASI ---
    ax.set_title('Top 10 Feature Extraction (TF-IDF)', 
                  fontweight='bold', fontsize=FONT_SIZE_TITLE, pad=30)
    ax.set_xlabel('Rata-rata Bobot (Average Weight)', fontweight='bold', fontsize=FONT_SIZE_AXIS_LABEL, labelpad=LABEL_PAD)
    ax.set_ylabel('Kata (Words)', fontweight='bold', fontsize=FONT_SIZE_AXIS_LABEL, labelpad=LABEL_PAD)

    # Menyesuaikan ukuran dan ketebalan label sumbu (ticks)
    ax.tick_params(axis='both', which='major', labelsize=FONT_SIZE_TICKS)
    plt.xticks(fontweight='bold') 
    plt.yticks(fontweight='bold') 

    plt.tight_layout()

    # --- 7. LOGIKA PENAMPILAN DAN PENYIMPANAN ---
    print("Membuka jendela visualisasi grafik...")
    plt.show()

    ans = input("Simpan grafik Top 10 TF-IDF dalam format SVG? (y/n): ").lower().strip()
    if ans == 'y':
        save_path = os.path.join(DIR_PATH, 'gambar15.svg')
        # Ekspor dalam format SVG untuk kualitas vektor maksimal
        fig.savefig(save_path, format='svg', bbox_inches='tight')
        print(f"Sukses: Grafik berhasil disimpan ke {save_path}")
    else:
        print("Grafik tidak disimpan.")

except Exception as e:
    print(f"Terjadi kesalahan: {e}")
```

## 10.9 Visualisasi 10 Fitur Teratas (Chi-Square)
Grafik ini menampilkan 10 kata dengan nilai statistik Chi-Square ($\chi^2$) tertinggi. Metrik ini menunjukkan kekuatan hubungan antara fitur (kata) dengan label target (kategori klasifikasi). Kata-kata yang muncul pada grafik ini merupakan fitur yang paling diskriminatif dan memiliki pengaruh paling signifikan bagi model dalam membedakan antara kelas Keluhan, Saran, dan Pujian.

```python
# --- LIBRARIES ---
import os
import pickle
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- 1. KONFIGURASI DIREKTORI DAN FILE ---
DIR_PATH = r'C:\Users\Asus\Desktop\SKRIPSI\1. MAIN'
MODEL_NAME = 'selector_chi2_2.0.pkl'
TFIDF_NAME = 'tfidf_model_2.0.pkl' 

# Konfigurasi Ukuran Font (Konsisten dengan format Skripsi)
FONT_SIZE_TITLE = 20
FONT_SIZE_AXIS_LABEL = 20
FONT_SIZE_TICKS = 16  
FONT_SIZE_BARS = 14  # Ukuran font untuk nilai skor di ujung batang
LABEL_PAD = 20
# -----------------------------------------------

try:
    # --- 2. PEMUATAN OBJEK SELECTOR DAN VECTORIZER ---
    with open(os.path.join(DIR_PATH, MODEL_NAME), 'rb') as f:
        selector = pickle.load(f)
    
    with open(os.path.join(DIR_PATH, TFIDF_NAME), 'rb') as f:
        tfidf = pickle.load(f)

    # --- 3. EKSTRAKSI NAMA FITUR DAN SKOR CHI-SQUARE ---
    features = tfidf.get_feature_names_out()
    scores = selector.scores_

    # --- 4. PENYUSUNAN DATAFRAME TOP 10 ---
    df_chi2 = pd.DataFrame({'word': features, 'score': scores})
    df_top10 = df_chi2.sort_values(by='score', ascending=False).head(10)

    # --- 5. PEMBUATAN GRAFIK (OBJECT-ORIENTED STYLE) ---
    sns.set_theme(style="whitegrid")
    fig, ax = plt.subplots(figsize=(12, 8)) 
    
    # Menggunakan palet 'rocket' untuk visualisasi gradasi yang kontras
    sns.barplot(
        data=df_top10, 
        x='score', 
        y='word', 
        palette='rocket', 
        edgecolor='black',
        ax=ax
    )

    # --- 6. FORMATTING TEKS DAN KETEBALAN ---
    ax.set_title('Top 10 Features Selection (Chi-Square)', 
                  fontweight='bold', fontsize=FONT_SIZE_TITLE, pad=30)
    ax.set_xlabel('Skor Chi-Square', fontweight='bold', fontsize=FONT_SIZE_AXIS_LABEL, labelpad=LABEL_PAD)
    ax.set_ylabel('Kata (Words)', fontweight='bold', fontsize=FONT_SIZE_AXIS_LABEL, labelpad=LABEL_PAD)
    
    # Menyesuaikan ukuran label sumbu (ticks)
    ax.tick_params(axis='both', which='major', labelsize=FONT_SIZE_TICKS)
    plt.xticks(fontweight='bold') 
    plt.yticks(fontweight='bold') 

    # --- 7. PENAMBAHAN LABEL SKOR DI UJUNG BATANG ---
    for p in ax.patches:
        score_val = p.get_width()
        ax.annotate(f'{score_val:.2f}', 
                    (score_val, p.get_y() + p.get_height() / 2.), 
                    ha='left', va='center', 
                    xytext=(8, 0), # Jarak teks dari ujung batang
                    textcoords='offset points', 
                    fontweight='bold', 
                    fontsize=FONT_SIZE_BARS)

    # Memberikan ruang tambahan di sisi kanan agar label skor tidak terpotong
    ax.set_xlim(0, df_top10['score'].max() * 1.15)
    plt.tight_layout()

    # --- 8. LOGIKA PENAMPILAN DAN PENYIMPANAN ---
    print("Membuka jendela visualisasi grafik...")
    plt.show()

    ans = input("Simpan grafik Chi-Square dalam format SVG? (y/n): ").lower().strip()
    if ans == 'y':
        save_path = os.path.join(DIR_PATH, 'gambar16.svg')
        # Ekspor dalam format SVG untuk kualitas vektor maksimal
        fig.savefig(save_path, format='svg', bbox_inches='tight')
        print(f"Sukses: Grafik berhasil disimpan ke {save_path}")
    else:
        print("Grafik tidak disimpan.")

except Exception as e:
    print(f"Terjadi kesalahan: {e}")
```

## 10.10 Visualisasi Koefisien Fitur Per Kelas
Grafik ini menampilkan 10 fitur (kata) dengan bobot koefisien tertinggi untuk masing-masing kelas target (Keluhan, Saran, dan Pujian). Analisis koefisien pada model *Logistic Regression* ini berfungsi untuk menjelaskan transparansi model (*model interpretability*), yaitu mengidentifikasi kata-kata kunci yang paling berkontribusi signifikan dalam mendorong keputusan klasifikasi model ke kelas tertentu.

```python
# --- LIBRARIES ---
import os
import pickle
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- 1. KONFIGURASI DIREKTORI DAN FILE ---
BASE_DIR = r'C:\Users\Asus\Desktop\SKRIPSI\1. MAIN'
MODEL_NAME = 'model_multinomial_lr_2.0.pkl'
TFIDF_NAME = 'tfidf_model_2.0.pkl'
SELECTOR_NAME = 'selector_chi2_2.0.pkl'
MAPPING_NAME = 'custom_mapping_2.0.pkl'

# Konfigurasi Ukuran Font (Konsisten dengan format Skripsi)
FONT_SIZE_MAIN_TITLE = 22
FONT_SIZE_SUB_TITLE = 18
FONT_SIZE_AXIS_LABEL = 16
FONT_SIZE_TICKS = 14  
LABEL_PAD = 20
# -----------------------------------------------

try:
    # --- 2. PEMUATAN MODEL DAN PENDUKUNG ---
    with open(os.path.join(BASE_DIR, MODEL_NAME), 'rb') as f:
        model = pickle.load(f)
    with open(os.path.join(BASE_DIR, TFIDF_NAME), 'rb') as f:
        tfidf = pickle.load(f)
    with open(os.path.join(BASE_DIR, SELECTOR_NAME), 'rb') as f:
        selector = pickle.load(f)
    with open(os.path.join(BASE_DIR, MAPPING_NAME), 'rb') as f:
        mapping = pickle.load(f)

    # Inversi mapping untuk keperluan tampilan label teks
    inv_mapping = {v: k for k, v in mapping.items()}

    # --- 3. PENYELARASAN NAMA FITUR ---
    all_features = tfidf.get_feature_names_out()
    # Mengambil hanya fitur yang lolos seleksi Chi-Square
    feature_names = all_features[selector.get_support()] 
    
    classes = model.classes_ 

    # --- 4. PEMBUATAN VISUALISASI ---
    sns.set_theme(style="whitegrid")
    # Membuat 1 baris dengan 3 kolom untuk masing-masing kelas (Keluhan, Saran, Pujian)
    fig, axes = plt.subplots(1, len(classes), figsize=(22, 10), sharex=False)

    for i, class_idx in enumerate(classes):
        # Mendapatkan label teks asli (Uppercase untuk judul sub-plot)
        label_text = inv_mapping.get(class_idx, str(class_idx)).upper()
        
        # Membuat DataFrame koefisien untuk setiap kelas
        coef_df = pd.DataFrame({'word': feature_names, 'coef': model.coef_[i]})
        top10 = coef_df.sort_values(by='coef', ascending=False).head(10)
        
        # Menggunakan palet 'viridis' untuk visualisasi yang profesional
        sns.barplot(data=top10, x='coef', y='word', ax=axes[i], palette='viridis', edgecolor='black')
        
        # Pengaturan Judul dan Label Sumbu per Sub-plot
        axes[i].set_title(f"Label: {label_text}", fontweight='bold', fontsize=FONT_SIZE_SUB_TITLE, pad=15)
        axes[i].set_xlabel('Bobot Koefisien', fontweight='bold', fontsize=FONT_SIZE_AXIS_LABEL, labelpad=LABEL_PAD)
        axes[i].set_ylabel('') # Nama kata sudah ada pada ticks sumbu Y

        # Mengatur ukuran dan ketebalan font pada sumbu (ticks)
        axes[i].tick_params(axis='both', which='major', labelsize=FONT_SIZE_TICKS)
        for label in (axes[i].get_xticklabels() + axes[i].get_yticklabels()):
            label.set_fontweight('bold')

    # Konfigurasi Judul Utama Grafik
    plt.suptitle('Top 10 Feature Coefficients per Class (Logistic Regression)', 
                 fontsize=FONT_SIZE_MAIN_TITLE, fontweight='bold', y=1.02)
    
    plt.tight_layout()

    # --- 5. LOGIKA PENAMPILAN DAN PENYIMPANAN ---
    print("Membuka jendela visualisasi grafik...")
    plt.show()

    ans = input("Simpan grafik Koefisien Model dalam format SVG? (y/n): ").lower().strip()
    if ans == 'y':
        save_path = os.path.join(BASE_DIR, 'gambar17.svg')
        # Ekspor dalam format SVG untuk kualitas vektor maksimal
        fig.savefig(save_path, format='svg', bbox_inches='tight')
        print(f"Sukses: Grafik koefisien berhasil disimpan ke {save_path}")
    else:
        print("Grafik tidak disimpan.")

except Exception as e:
    print(f"Terjadi kesalahan: {e}")
```

## 10.12 Visualisasi Confusion Matrix dan Laporan Klasifikasi
Tahap akhir visualisasi ini menampilkan *Confusion Matrix* untuk mengevaluasi sebaran prediksi model terhadap label aktual secara mendetail. Grafik didesain dengan format *compact* dan diekspor ke dalam format SVG untuk menjaga kualitas vektor. Selain itu, sistem akan menampilkan tabel laporan klasifikasi (*Classification Report*) yang mencakup metrik *Precision*, *Recall*, dan *F1-Score* untuk setiap kelas.

```python
# --- LIBRARIES ---
import os
import pickle
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import sparse
from sklearn.metrics import classification_report, confusion_matrix

# --- 1. KONFIGURASI DIREKTORI DAN PATH ---
BASE_DIR = r'C:\Users\Asus\Desktop\SKRIPSI\1. MAIN'

# Konfigurasi Teks dan Font (Konsisten dengan format Skripsi)
PLOT_TITLE = 'Confusion Matrix'
X_LABEL = 'Label Prediksi (Predicted)'
Y_LABEL = 'Label Aktual (Actual)'
SAVE_NAME = 'gambar19.svg'  

FONT_SIZE_TITLE = 20
FONT_SIZE_AXIS_LABEL = 20
FONT_SIZE_TICKS = 16  
FONT_SIZE_BARS = 16  # Ukuran angka di dalam matriks
LABEL_PAD = 20
# -----------------------------------------------

try:
    # --- 2. PEMUATAN MODEL, MAPPING, DAN DATA ---
    with open(os.path.join(BASE_DIR, 'model_multinomial_lr_2.0.pkl'), 'rb') as f:
        model = pickle.load(f)
    with open(os.path.join(BASE_DIR, 'custom_mapping_2.0.pkl'), 'rb') as f:
        custom_mapping = pickle.load(f)

    X_test_selected = sparse.load_npz(os.path.join(BASE_DIR, 'X_test_selected_2.0.npz'))
    test_df = pd.read_csv(os.path.join(BASE_DIR, 'data_test_final_2.0.csv'))

    # --- 3. PROSES PREDIKSI ---
    y_true = test_df['label_pks'].map(custom_mapping)
    y_pred = model.predict(X_test_selected)
    
    # Inversi mapping untuk label sumbu
    reverse_mapping = {v: k for k, v in custom_mapping.items()}
    target_names = [f"{reverse_mapping[i]} ({i})" for i in sorted(reverse_mapping.keys())]

    # --- 4. VISUALISASI: COMPACT CONFUSION MATRIX ---
    cm = confusion_matrix(y_true, y_pred)
    
    sns.set_theme(style="white") 
    fig, ax = plt.subplots(figsize=(8, 8))
    
    # Heatmap dengan skema warna 'Blues' dan bentuk sel persegi sempurna (square=True)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', square=True,
                xticklabels=target_names, yticklabels=target_names, 
                cbar_kws={"shrink": .8}, ax=ax,
                annot_kws={"size": FONT_SIZE_BARS, "fontweight": "bold"})
    
    # Penerapan Tipografi dan Spasi
    ax.set_title(PLOT_TITLE, fontsize=FONT_SIZE_TITLE, fontweight='bold', pad=30)
    ax.set_xlabel(X_LABEL, fontsize=FONT_SIZE_AXIS_LABEL, fontweight='bold', labelpad=LABEL_PAD)
    ax.set_ylabel(Y_LABEL, fontsize=FONT_SIZE_AXIS_LABEL, fontweight='bold', labelpad=LABEL_PAD)
    
    # Mengatur ketebalan dan ukuran tick labels
    ax.tick_params(axis='both', which='major', labelsize=FONT_SIZE_TICKS)
    plt.xticks(fontweight='bold') 
    plt.yticks(fontweight='bold', rotation=0) 

    plt.tight_layout()

    print("Menampilkan visualisasi Confusion Matrix...")
    plt.show()

    # Logika Penyimpanan
    ans = input(f"Simpan grafik Confusion Matrix sebagai {SAVE_NAME}? (y/n): ").lower().strip()
    if ans == 'y':
        save_path = os.path.join(BASE_DIR, SAVE_NAME)
        fig.savefig(save_path, format='svg', bbox_inches='tight')
        print(f"Sukses: Grafik berhasil disimpan ke {save_path}")

    # --- 5. TABEL CLASSIFICATION REPORT ---
    print("\nLAPORAN KLASIFIKASI (CLASSIFICATION REPORT)")
    print("-" * 60)
    report_dict = classification_report(y_true, y_pred, target_names=target_names, output_dict=True)
    report_df = pd.DataFrame(report_dict).transpose()
    print(report_df.round(2).to_string())
    print("-" * 60)

except Exception as e:
    print(f"Terjadi kesalahan: {e}")
```

Visualisasi diagram batang berkelompok (*grouped bar chart*) ini menampilkan perbandingan metrik evaluasi yang meliputi *Precision*, *Recall*, dan *F1-Score* untuk setiap kategori label. Grafik ini memberikan gambaran komprehensif mengenai efektivitas model dalam mengklasifikasikan masing-masing kelas, serta menggunakan skema warna *matte* yang seragam untuk menjaga konsistensi visual di seluruh laporan penelitian.

```python
# --- LIBRARIES ---
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- 1. KONFIGURASI DIREKTORI ---
BASE_DIR = r'C:\Users\Asus\Desktop\SKRIPSI\1. MAIN'

# Konfigurasi Ukuran Font (Konsisten dengan format Skripsi)
FONT_SIZE_TITLE = 20
FONT_SIZE_AXIS_LABEL = 20
FONT_SIZE_TICKS = 16  
FONT_SIZE_BARS = 14  # Ukuran label persentase di atas batang
LABEL_PAD = 20
# -----------------------------------------------

# --- 2. PERSIAPAN DATA ---
# Data diekstrak dari tabel hasil evaluasi dan dikonversi ke format persentase
data = {
    'Label': ['Keluhan (0)', 'Saran (1)', 'Pujian (2)'],
    'Precision': [0.90 * 100, 0.49 * 100, 0.76 * 100],
    'Recall': [0.80 * 100, 0.64 * 100, 0.85 * 100],
    'F1-Score': [0.84 * 100, 0.56 * 100, 0.80 * 100]
}

df_metrics = pd.DataFrame(data)
df_melted = df_metrics.melt(id_vars='Label', var_name='Metric', value_name='Percentage')

try:
    # --- 3. PENGATURAN TEMA VISUAL ---
    sns.set_theme(style="whitegrid")
    fig, ax = plt.subplots(figsize=(14, 8)) 

    # PEMETAAN WARNA MATTE: Merah (Keluhan), Kuning (Saran), Hijau (Pujian)
    custom_palette = ['#D65A5A', '#E6C95D', '#7FB285']

    # Pembuatan Grouped Bar Chart
    sns.barplot(data=df_melted, x='Metric', y='Percentage', hue='Label', 
                palette=custom_palette, edgecolor='black', ax=ax, width=0.7)

    # --- 4. ANOTASI DAN FORMATTING ---
    ax.set_title('Model Performance Comparison per Class', fontsize=FONT_SIZE_TITLE, fontweight='bold', pad=30)
    ax.set_ylabel('Persentase (%)', fontsize=FONT_SIZE_AXIS_LABEL, fontweight='bold', labelpad=LABEL_PAD)
    ax.set_xlabel('Metrik Evaluasi', fontsize=FONT_SIZE_AXIS_LABEL, fontweight='bold', labelpad=LABEL_PAD)

    # Mengatur ukuran dan ketebalan label sumbu (ticks)
    ax.tick_params(axis='both', which='major', labelsize=FONT_SIZE_TICKS)
    plt.xticks(fontweight='bold') 
    plt.yticks(fontweight='bold') 

    # Menambahkan label persentase tebal di atas setiap batang
    for p in ax.patches:
        height = p.get_height()
        if height > 0:
            ax.annotate(f'{int(height)}%', 
                        (p.get_x() + p.get_width() / 2., height),
                        ha='center', va='bottom', 
                        xytext=(0, 8), 
                        textcoords='offset points', 
                        fontweight='bold', fontsize=FONT_SIZE_BARS)

    # Mengatur batas sumbu Y untuk ruang anotasi
    ax.set_ylim(0, 115) 

    # Konfigurasi Legenda
    plt.legend(title='Kategori', bbox_to_anchor=(1.02, 1), loc='upper left', 
               fontsize=12, title_fontsize=13, frameon=True)
    
    plt.tight_layout()

    # --- 5. TAHAP PENAMPILAN DAN PENYIMPANAN ---
    print("Membuka jendela visualisasi grafik perbandingan performa...")
    plt.show()

    ans = input("Simpan grafik perbandingan performa dalam format SVG? (y/n): ").lower().strip()
    if ans == 'y':
        save_path = os.path.join(BASE_DIR, 'gambar20.svg')
        # Ekspor secara eksplisit ke format SVG untuk kualitas vektor maksimal
        fig.savefig(save_path, format='svg', bbox_inches='tight')
        print(f"Sukses: Grafik berhasil disimpan ke {save_path}")
    else:
        print("Penyimpanan dibatalkan.")

except Exception as e:
    print(f"Terjadi kesalahan: {e}")
```
